````python
import matplotlib.pyplot as plt
import random
````

````python
r = 0
population = 0.5
````

The equation is of the form x_n = A * x_(n - 1) * (1 - x_(n - 1))

```itr``` is the number of iterations.
```p``` is the percentage of population.

````python
pop = []
growth = []
````

````python
def inc(p):
    return round(r * p * (1 - p), 3)
````

````python
while (r < 5):
    p = population
    p_values = []
    coll = []
    for i in range(1000):
        p_values.append(p)
        p = inc(p)
    for i in p_values:
        if i not in coll and p_values.count(i) > 50:
            coll.append(i)
    for i in range(len(coll)):
        pop.append(coll[i])
        growth.append(r)
    r += 0.001
````

````python
fig, ax = plt.subplots()
ax.plot(growth, pop, ".", markersize=0.2)
# ax.axis('equal')
ax.set_xlabel("r")
ax.set(ylim=(0, 1))
````
